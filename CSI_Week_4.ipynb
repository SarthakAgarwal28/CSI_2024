```python
# WAP to normalize Iris Dataset
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()

print("Original Iris Dataset:")
print(iris.data)

iris_features = iris.data

def x_normalized(x, a, b):
    x_min = np.min(x)
    x_max = np.max(x)
    normalized_x = a + (2*((x - x_min) / (x_max - x_min)))
    return normalized_x

a = -1
b = 1

iris_normalized_features = np.apply_along_axis(x_normalized, 0, iris_features, a=a, b=b)

print("\nNormalized Iris Dataset:")
print(iris_normalized_features)
```

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Load Iris dataset
iris = load_iris()
data = iris.data
target = iris.target

# Convert dataset to numpy array
data_np = np.array(data)
target_np = np.array(target)

# Find features, classes, and examples
num_features = data_np.shape[1]
num_classes = len(np.unique(target_np))
num_examples = len(data_np)

# Min-Max normalization
#min_max_scaler = MinMaxScaler()
data_min_max = MinMaxScaler().fit_transform(data_np)

# Z-score normalization
#z_score_scaler = StandardScaler()
data_z_score = StandardScaler().fit_transform(data_np)

# Covariance Matrix
covariance_matrix = np.cov(data_np.T)

# Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Print results
print("Number of features:", num_features)
print("Number of classes:", num_classes)
print("Number of examples:", num_examples)
print("Min-Max normalized data:\n", data_min_max)
print("Z-score normalized data:\n", data_z_score)
print("Covariance Matrix:\n", covariance_matrix)
print("Eigenvalues:\n", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```
